```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# FPT


```python
FPT=pd.read_csv('/content/FPT.csv',sep=";")
FPT
```

```python
# 3.1 Trung bình giá trị GDP
FPT['Price'].mean()
# 3.2 Tính trung vị giá trị GDP
FPT['Price'].median()
# 3.3 Tìm Max/Min giá trị GDP
print(FPT['Price'].max())
print(FPT['Price'].min())
# 3.4 Thống kê nhanh giá trị GDP
FPT['Price'].describe()
```

```python
# 3.14 Vẽ Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=FPT['Price'], color='#03dffc', width=0.5, linewidth=2)
plt.title('Box Plot of FPT Stock Price')
plt.xlabel('Price')
plt.show()
```

```python
# Vẽ Histogram
plt.figure(figsize=(10, 6))
sns.histplot(FPT['Price'],color='#03dffc', bins=20, kde=True)
plt.title('Histogram of FPT Stock Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```
